# Production Scheduling Optimization Demo

This notebook demonstrates how to analyze the summary results for the Genetic Algorithm production scheduling project.

It covers:

- Benchmark results
- Average deviation from optimal
- Mathematical model vs GA time comparison
- Number of times the optimal value was found


```
import pandas as pd
import matplotlib.pyplot as plt

benchmarks = pd.read_csv("../data/benchmark_instances.csv")
ga = pd.read_csv("../data/ga_results_summary.csv")
time_comp = pd.read_csv("../data/mathematical_model_vs_ga_time.csv")

ga.head()
```

## Average deviation from optimal

```
plt.figure(figsize=(12, 5))
plt.bar(ga["problem"], ga["percent_average_deviation_from_optimal"])
plt.title("Average deviation from optimal")
plt.ylabel("Deviation (%)")
plt.xticks(rotation=65)
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

## Solution time comparison

```
plt.figure(figsize=(12, 5))
plt.plot(time_comp["problem"], time_comp["mathematical_model_time_seconds"], marker="o", label="Mathematical model")
plt.plot(time_comp["problem"], time_comp["proposed_ga_time_seconds"], marker="o", label="Proposed GA")
plt.title("Solution time comparison")
plt.ylabel("Seconds")
plt.xticks(rotation=65)
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()
```

## Times optimal was found

```
plt.figure(figsize=(12, 5))
plt.bar(ga["problem"], ga["times_optimal_found_out_of_100"])
plt.title("Times the proposed GA found the optimal value")
plt.ylabel("Times out of 100 runs")
plt.xticks(rotation=65)
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

## Key interpretation

The proposed GA produces strong results for most benchmark instances. It is especially useful as a practical decision-support approach when exact mathematical models become slow or difficult to solve at scale.
